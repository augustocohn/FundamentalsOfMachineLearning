2. Normalize the selected feature using the min-max scaling method. Let $x_1$ be one of the selected feature:

$$\hat{x_1} = \frac{x_1 - \min(x_1)}{\max(x_1) - \min(x_1)}$$

where $\max(x_1)$ is the maximum value of feature $x_1$ and $\min(x_1)$ is the minimum value of feature $x_1$.

````python
x = df['LSTAT'].to_numpy()
x_hat = (x - np.min(x))/(np.max(x) - np.min(x))
t = df['House Value'].to_numpy()
````

3. Implement a polynomial regression model using 4-fold cross-validation to select the best model order $M$. Use the coefficient of determination, $r^2$, as the goodness measure.

You should **not** use built-in functions to implement polynomial regression or cross-validation. But you can use help functions such as [```train_test_split```](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) to split the data, [```linregress```](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html) to compute the coefficient of determination and [```KFold```](https://scikit-learn.org/0.16/modules/generated/sklearn.cross_validation.KFold.html) to split the training set in k-folds.

* partition the (scaled) data into training (80%) and test (20%) sets
* use 4-fold cross-validation to select a best value for model order $M$ from a pre-defined set
* train a final model with the *best* model order
* evaluate performance in the test set using the same goodness measure ($r^2$)
* draw conclusions

````python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from scipy.stats import linregress
````

````python
def PolynomialRegression(x,t,M): 
    #COMPUTER FEATURE MATRIX
    X = np.array([x**m for m in range(M+1)]).T
    #COMPUTER COEFFICIENTS w
    w = np.linalg.inv(X.T@X)@X.T@t
    #MODEL PREDICTION
    y = X@w
    #RESIDUAL
    err = t-y
    return w, y, err

def PolynomialRegression_test(x, w, M):
    X = np.array([x**m for m in range(M+1)]).T
    y = X@w 
    return y
````

````python
x_train, x_test, t_train, t_test = train_test_split(x_hat, t, test_size = 0.2)
r_sq_train_ar, r_sq_val_ar= [], []
````

````python
Mrange = 16
k_folds = 4
````

````python
kf = KFold(n_splits=k_folds, shuffle=True)
````

````python
for M in range(1, Mrange):
    print('M value: ', M)
    f = 1
    r_sq_train_avg, r_sq_val_avg = 0, 0
    for train_i, val_i in kf.split(x_train):
        print('\nFold: ', f)
        
        x_train_fold = x_train[train_i]
        t_train_fold = t_train[train_i]
        x_val_fold = x_train[val_i]
        t_val_fold = t_train[val_i]
        
        w, y_train, _  = PolynomialRegression(x_train_fold, t_train_fold, M)
        
        y_val = PolynomialRegression_test(x_val_fold, w, M)
        
        _, _, r_train, _, _ = linregress(np.sort(t_train_fold), np.sort(y_train))
        _, _, r_val, _, _ = linregress(np.sort(t_val_fold), np.sort(y_val))
        
        r_sq_train_avg += r_train**2
        r_sq_val_avg += r_val**2
        
        print('R Sq Training: ', r_train**2)
        print('R Sq Validation: ', r_val**2)
        
        f += 1
        
    print('\nAverage R Sq Training: ', r_sq_train_avg/k_folds)
    print('Average R Sq Validation: ', r_sq_val_avg/k_folds)
    print('-------------------------------------------------')
    r_sq_val_ar.append(r_sq_val_avg/k_folds)
    r_sq_train_ar.append(r_sq_train_avg/k_folds)
````

````python
r_sq_val_ar = np.array(r_sq_val_ar)
M_best = np.argmax(r_sq_val_ar)+1
print('Max R Sq avg: ', np.max(r_sq_val_ar), 'at M = ', M_best)
````

````python
w, y_train, _ = PolynomialRegression(x_train, t_train, M_best)
y_test = PolynomialRegression_test(x_test, w, M_best)

#ALLOWS FOR MODEL TO EXTEND THE FULL LENGTH OF THE PLOT
y_full = PolynomialRegression_test(x_hat, w, M_best)

_, _, r, _, _ = linregress(np.sort(t_test), np.sort(y_test))

plt.scatter(x_train, t_train, label='Training Data')
plt.scatter(x_test, t_test, c='b', marker='*', label='Test Data')
# plt.plot(np.sort(x_test), np.sort(y_test)[::-1], c ='r', label='Model')
plt.plot(np.sort(x_hat), np.sort(y_full)[::-1], c ='r', label='Model')
plt.title('Regression for LSTAT and House Value')
plt.ylabel('House Value')
plt.xlabel('LSTAT (scaled)')
plt.legend(fontsize=10)
print('R Sq for best model order: ', r**2)
````

When rerunning the training, I was getting a very wide range of possible best M values (I saw between 2 - 13). I assume this is due to the small dataset that we are working with. When the data is shuffled in performing k-fold cross-validation, the resulting model produced very different results. With a much larger dataset, there would be a much smaller range of M values produced when performing k-fold and with a large enough dataset, the best M might converge to one value. Since there was a large range of M values for this dataset, Occam's razor says we should select a simpler model order if it provides a similar error. Below is an analysis on the $R^2$ for model orders within 3 of the selected best model order. 

````python
for M in range(M_best-3, M_best+4):
    print(f'For M = {M}: {r_sq_val_ar[M-1]}')
````

As we can see, the $R^2$ values for each model order are very similar and while $M_{best}$ provides the largest correlation, a more simple model can provide the same correlation up to the ten-thousandths place.
